```python
# Função / Método	O que faz
# df.head(n)	Mostra as primeiras n linhas
# df.tail(n)	Mostra as últimas n linhas
# df.info()	Mostra infos gerais (tipos, valores nulos, etc.)
# df.describe()	Estatísticas descritivas para colunas numéricas
# df['coluna'] ou df.coluna	Acessa uma coluna específica
# df.loc[linha]	Acessa por rótulo (nome do índice)
# df.iloc[linha]	Acessa por posição
# df[df['coluna'] == valor]	Filtra linhas com base em uma condição
# df[df['coluna'].str.contains('texto')]	Filtra por texto parcial (em strings)
# df.drop(columns=['coluna1', 'coluna2'])	Remove colunas
# df.rename(columns={'antigo': 'novo'})	Renomeia colunas
# df.sort_values('coluna')	Ordena por coluna
# df.groupby('coluna')	Agrupa valores iguais para análises
# df.isnull() / df.dropna() / df.fillna()	Trabalha com valores nulos
```

```python
import pandas  as pd

df = pd.read_csv('imoveis_ficticios.csv', sep=',')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#NOVO DF APARTAMENTOS
df_Apartamentos = df[df['Tipo_Imovel'] == 'Apartamento']
df_Apartamentos.head()
```

```python
#CALCULAR O VALOR DO M2
# valor_m2 = valor_aluguel / area_em_m2

df_Apartamentos.loc[:,'Valor_m2'] = df_Apartamentos['Aluguel_Mensal'] / df_Apartamentos['Area_m2']
df_Apartamentos.head()
```

```python
#ORDENAR POR VALOR M2
df_Apartamentos.sort_values(by='Valor_m2', ascending=False).head(10)
```

```python
#Arrendondar os valores dos m²
df_Apartamentos.loc[:,'Valor_m2'] = df_Apartamentos['Valor_m2'].round(0)
```

```python
#funcao de alta demanda
media_geral = df_Apartamentos['Valor_m2'].mean()
media_geral.round(0)

# print(f'A média geral dos m² é de R${media_geral:.2f}')

def alta_demanda(valor_m2):
    if valor_m2 > media_geral:
        return'Acima da média'
    if valor_m2 == media_geral:
        return 'Dentro da média'
    else:
        return 'Abaixo da média'
```

```python
df_Apartamentos.loc[:,'Alta_demanda'] = df_Apartamentos['Valor_m2'].apply(alta_demanda)

df_Apartamentos.head()
```

```python
df_Apartamentos.to_csv('Apartamentos_Alta_Demanda.csv')
```
